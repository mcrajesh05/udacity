```python
%sql SELECT a.*,  s.*\
FROM  sparkify.artists a  LEFT JOIN  sparkify.songs s\
ON a.artist_id = s.artist_id
```

```python
%sql SELECT DISTINCT  u.first_name, u.last_name, s.title, sp.location, sp.user_agent\
FROM sparkify.users u \
INNER JOIN sparkify.songplays sp ON u.user_id = sp.user_id\
INNER JOIN sparkify.songs s ON sp.song_id = s.song_id
```
